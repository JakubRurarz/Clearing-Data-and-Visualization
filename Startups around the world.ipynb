```python
kraje = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
iso = dane_poczatkowe.groupby('country_code').size().reset_index(name='counts')
razem = kraje.merge(iso, left_on='iso_a3', right_on='country_code')
plt.style.use('dark_background')
razem.plot(column='counts', cmap='OrRd', legend=True, figsize=(20,12))
plt.title('Startupy na swiecie')
plt.show()
```

```python
razem_bez_usa = razem.iloc[np.where(razem['country_code'] != 'USA')]
razem_bez_usa.plot
razem_bez_usa.plot(column='counts', cmap='OrRd', legend=True, figsize=(20,12))
plt.title('startupy na swiecie w krajach poza USA')
plt.show()
```

```python
na = dane_poczatkowe['category_list'].isna().sum()
branza = dane_poczatkowe['category_list']
branza = branza.fillna('Other')
branza.value_counts()
```

```python
branza_normalizacja = branza.value_counts(normalize=True)
top_1_procent = branza_normalizacja[branza_normalizacja>0.01]
top_1_procent.plot(kind='barh')
plt.title('Top 1% branzy dla startupow')
plt.xlabel('czestotliwosc wystepowania')
plt.grid(True)
plt.text(0.8*max(top_1_procent), len(top_1_procent)-1,f'Suma: {top_1_procent.sum().round(3)*100}%', ha='left', va='top',color='white', bbox=dict(facecolor='blue', alpha=1))
plt.show()
```

```python
udane = dane_poczatkowe.loc[dane_poczatkowe['status'] == 'closed']
na2 = dane_poczatkowe['category_list'].isna().sum()
```

```python
branza_udane = udane['category_list']
branza_udane = branza_udane.fillna('Other')
branza_udane.value_counts()
```

```python
branza_normalizacja_udane = branza_udane.value_counts(normalize=True)
top_1_procent_udane = branza_normalizacja_udane[branza_normalizacja_udane>0.01]
top_1_procent_udane.plot(kind='barh')
plt.title('Top 1% najczesciej upadajacych branzy dla startupow')
plt.xlabel('czestotliwosc wystepowania')
plt.grid(True)
plt.text(0.8*max(top_1_procent_udane), len(top_1_procent_udane)-1,f'Suma: {top_1_procent_udane.sum().round(3)*100}%', ha='left', va='top',color='white', bbox=dict(facecolor='blue', alpha=1))
plt.show()
```
